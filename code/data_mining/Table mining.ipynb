```python
import pandas as pd
import yfinance as yf
import datetime
```

```python
def yfinance_ask(company):
    company = company
    start_date = '1920-01-01'
    end_date = (datetime.datetime.now()).strftime("20%y-%m-%d")
    interval = '1d'
    print(f'Download information about {company}, from {start_date} to {end_date} with interval {interval}')
    data = yf.download(company, start=start_date, end=end_date, interval=interval)
    return data
```

```python
data = pd.DataFrame(columns=['Date','Ticker','Open', 'Close','Traiding_result'])
company = 'JNJ'
```

```python
company_actions_traiding_story = yfinance_ask(company)
```

```python
company_actions_traiding_story = company_actions_traiding_story.reset_index()
company_actions_traiding_story
```

```python
data['Date'] = company_actions_traiding_story['Date']
data['Ticker'] = str(company)
data['Open'] = company_actions_traiding_story['Open']
data['Close'] = company_actions_traiding_story['Close']
data['Traiding_result'] = data.apply(lambda row: 1 if row['Open'] < row['Close'] else 0, axis = 1)
data.set_index('Date', inplace=True)
data
```

```python
data.to_csv(f'Traiding_table_{company}.csv', sep=',', index=True)
```
