```python
import pandas as pd
import numpy as np
import ast
import warnings
warnings.filterwarnings('ignore')
```

```python
AAPL = pd.read_csv("data_unprocessed/AAPL_data.csv").set_index('Date').loc['2021-01-05':]
AMZN = pd.read_csv("data_unprocessed/AMZN_data.csv").set_index('Date').loc['2021-01-05':]
GOOGL = pd.read_csv("data_unprocessed/GOOGL_data.csv").set_index('Date').loc['2021-01-05':]
JNJ = pd.read_csv("data_unprocessed/JNJ_data.csv").set_index('Date').loc['2021-01-05':]
df = pd.concat([AAPL, AMZN, GOOGL, JNJ], axis=0)
df
```

```python
columns_sentiments = []
columns_neg = []
columns_neu = []
columns_pos = []
for column in df.columns:
    column_sentiment = []
    column_neg = []
    column_neu = []
    column_pos = [] 
    if column.startswith('sentiment'):
        for sentiment_dict in df[column]:
            try:
                sentiment_dict = ast.literal_eval(sentiment_dict)
                column_sentiment.append(sentiment_dict['polarity'])
                column_neg.append(sentiment_dict['neg'])
                column_neu.append(sentiment_dict['neu'])
                column_pos.append(sentiment_dict['pos'])
            except Exception as e:
                column_sentiment.append(np.nan)
                column_neg.append(np.nan)
                column_neu.append(np.nan)
                column_pos.append(np.nan)
        columns_sentiments.append(column_sentiment)
        columns_neg.append(column_neg)
        columns_neu.append(column_neu)
        columns_pos.append(column_pos)
```

```python
# Creating average polarity per day
daily_polarity = []
daily_neg = []
daily_neu = []
daily_pos = []
daily_cols_lists = ['columns_sentiments','columns_neg','columns_neu','columns_pos']
daily_cols_means = ['daily_polarity','daily_neg','daily_neu','daily_pos']

for i in range(len(daily_cols_lists)):
    exec(f'matrix = np.array({daily_cols_lists[i]}).transpose()')
    for day_sentiment in matrix:
        exec(f'{daily_cols_means[i]}.append(np.mean([x for x in day_sentiment if str(x)!="nan"]))')

    exec(f'df["{daily_cols_means[i]}"] = {daily_cols_means[i]}')
```

```python
# Deleting unnecessary columns

main_data = df.copy()
for column in main_data.columns:
    if column.startswith('news') or column.startswith('sentiment') or column.startswith('Ticker'):
        del main_data[column]
main_data = main_data.reset_index()
main_data['Date'] = pd.to_datetime(main_data['Date'])
main_data_X = main_data.groupby(pd.Grouper(key='Date', freq='W')).mean()

def calculate_mode(series):
    mode_result = series.mode()
    if not mode_result.empty:
        return mode_result.iloc[0]
    else:
        return None
main_data_Y = main_data.groupby(pd.Grouper(key='Date', freq='W')).agg(calculate_mode)
main_data.reset_index()['Date']
del main_data_X['Trading_result']
main_data_X['Trading_result'] = [int(x) if str(x)!='nan' else main_data_Y['Trading_result'].mode() for x in main_data_Y['Trading_result']]
main_data = main_data_X
main_data['Volume'] = [x if str(x)!='nan' else main_data['Volume'].mean() for x in main_data['Volume']]
main_data['daily_polarity'] = [x if str(x)!='nan' else main_data['daily_polarity'].mean() for x in main_data['daily_polarity']]
main_data['daily_neg'] = [x if str(x)!='nan' else main_data['daily_neg'].mean() for x in main_data['daily_neg']]
main_data['daily_neu'] = [x if str(x)!='nan' else main_data['daily_neu'].mean() for x in main_data['daily_neu']]
main_data['daily_pos'] = [x if str(x)!='nan' else main_data['daily_pos'].mean() for x in main_data['daily_pos']]
main_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_validate
```

```python
# Testing with and without different columns

main_data_test = main_data.copy()
del main_data_test['daily_polarity']
#del main_data_test['daily_neg']
#del main_data_test['daily_neu']
#del main_data_test['daily_pos']
#del main_data_test['Volume']


# Creating trading_result for next day
main_data_test['Trading_result_next_day'] = [0] + list(main_data_test['Trading_result'])[:-1]

del main_data_test['Trading_result']
main_data_test
```

```python
X = main_data_test.drop(['Trading_result_next_day'],axis=1)
Y = main_data_test['Trading_result_next_day'].astype(int)
```

```python
# Checking ratio of target values
target_values = [len(Y[Y==0]),len(Y[Y==1])]
labels = ['Zero','One']
palette_color = sns.color_palette('Set2')
plt.pie(target_values,colors=palette_color,labels=labels)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
# Creating different models
lgr=LogisticRegression(max_iter = 1000)
bern = BernoulliNB()
rfc = RandomForestClassifier(n_estimators=200, max_depth=3)
tree = tree.DecisionTreeClassifier() #high test_score, the roc_auc is needed to be checked
models = ['rfc','bern','tree','lgr']
```

```python
# Finding the best model with cross-validation, detecting overfitting by checking test_scores and fit_time
print('Finding the best model with cross-validation, detecting overfitting by checking test_scores and fit_time')
long_check = []
for model in models:
    count = 0
    model_hundred_results = []
    while count<10:
        exec(f'smth = cross_validate({model},x_train,y_train,cv=3)')
        model_hundred_results.append(smth['test_score'])
        count+=1
    long_check.append(model_hundred_results)
```

```python
for results in long_check:
    print(np.mean(np.array(results)))
```

```python
print("Tree model showed the best test_score, maybe it's the best model, but to be precise I will check roc_auc score")
for model in models:
    exec(f'{model}.fit(x_train,y_train)')
for model in models:
    exec(f'print(roc_auc_score(y_test, {model}.predict_proba(x_test)[:, 1]))')
```
